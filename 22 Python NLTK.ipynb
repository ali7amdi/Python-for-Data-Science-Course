```python
from nltk.corpus import wordnet
```

```python
syns = wordnet.synsets("program")
```

```python
syns
```

```python
syns[0].lemmas()[0].name()
```

```python
syns[4].definition()
```

```python
syns[0].examples()
```

```python
synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())
```

```python
synonyms
```

```python
print (antonyms)
```
